```python
import pandas as pd
```

```python
vine_review_df = pd.read_csv('vine_table.csv')
vine_review_df.head(12)
```

```python
# 1. Filter the data and create a new DataFrame or table to retrieve
# all the rows where the total_votes count is equal to or greater than 20
```

```python
votes_vine_df = vine_review_df[vine_review_df['total_votes'] >= 20]
votes_vine_df.head(12)
```

```python
# 2. Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve
# all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
```

```python
helpful_vine_df = votes_vine_df.loc[votes_vine_df.helpful_votes / votes_vine_df.total_votes >= 0.5]
helpful_vine_df.head(12)
```

```python
# 3. Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves 
# all the rows where a review was written as part of the Vine program (paid
```

```python
reviews_written = helpful_vine_df[helpful_vine_df['vine'] == 'Y']
reviews_written.head(12)
```

```python
# Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid)
```

```python
not_paid_reviews = helpful_vine_df[helpful_vine_df['vine'] == 'N']
not_paid_reviews.head(12)
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
```

```python
paid_reviews_written_total = len(reviews_written)
paid_reviews_written_total
```

```python
not_paid_reviews_total = len(not_paid_reviews)
not_paid_reviews_total
```

```python
five_star_total_paid = len(reviews_written[reviews_written['star_rating']==5])
five_star_total_paid
```

```python
five_star_total_not_paid = len(not_paid_reviews[not_paid_reviews['star_rating']==5])
five_star_total_not_paid
```

```python
# Percentage of 5-star reviews (paid)
```

```python
five_star_total_paid/paid_reviews_written_total
```

```python
# Percentage of 5-star reviews (not paid)
```

```python
five_star_total_not_paid/not_paid_reviews_total
```
